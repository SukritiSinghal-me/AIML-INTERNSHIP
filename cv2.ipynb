```python
import numpy as np
import matplotlib.pyplot as plt

# Create a blank image (height, width, channels)
image = np.full((18, 24, 3), 255, dtype=np.uint8)

# Top band - Saffron
image[0:6, :, :] = [255, 153, 51]

# Middle band - White
image[6:12, :, :] = [255, 255, 255]

image[9,12] = [0,0,128]

# Bottom band - Green
image[12:18, :, :] = [19, 136, 8]

plt.imshow(image)
plt.axis('off')
plt.show()
```
